# Creating the Saichev and Sorenette equation 

This equation takes the Gutenberg-Richter law, Omori law and productivity law and creates a probabily density function of the interevent recurrence times

Here we see the equation outlined which depcits the probability denisty function for our recurrnce times of an earthquake.

```python
#importing packages

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
import plotly.express as px
from IPython.display import display
import ipywidgets as widgets 

```

```python
# Denfining x and our plot with minium, maximum and step size

x = np.arange(0.001, 10, 0.001)


# Defining the functions paramters


a = 0.76
n = 0.8
θ = 0.03
ρ = 1


#a = widgets.FloatSlider(min = 0.1, max = 1, value =  0.76, description = '....')
#n = widgets.FloatSlider(min = 0.1, max = 1, value =  0.8, description = '....')
#θ = widgets.FloatSlider(min = 0.1, max = 1, value = 0.03, description = '....')

# Saichev and Sorenette created a range of p from 1 to 10^-3



# defining the function 

y = ((a*n*θ*ρ**(θ)*x**(-1-θ)) + (1-n+a*n*ρ**(θ)*(x**-θ))**2) * np.exp(-(1-n)*x-(a*n*(ρ**θ)/(1-θ))*x**(1-θ))

# defining the power laws 

y2 = x**(-1) / 30
y3 = x**(-0.3) / 3


# changing size of the figure  

plt.figure(figsize=(10, 10))

# Labelling the functions plot

plt.title("Saichev and Sorenette Equation 11")
plt.xlabel("Values of  x")
plt.ylabel("Values of f(x)")  

# Changing the range 

plt.ylim([0.0001, 100])
plt.xlim([0.001, 10])

# Plotting the function

plt.yscale("log")
plt.xscale("log")
plt.plot(x,y, color = 'b' ,label = "main function" )
plt.plot(x,y2, '--' , color = 'g' ,label = "x^-1")
plt.plot(x,y3, '--' , color = 'm' ,label = "x^-0.3")
plt.legend() 
plt.show()
plt.savefig("Saichev and Sornette")



```

```python
# Denfining x and our plot with minium, maximum and step size and parameters for function 

def fn(A = 0.76 , N = 0.80 ,Theta = 0.03, Rho = 1):
    x = np.arange(0.00001, 10, 0.00001)


    # Defining the functions paramters and widgets


    a = A
    n = N
    θ = Theta
    ρ = Rho
    
    aconstant = 0.76
    nconstant = 0.8
    θconstant = 0.03
    

    # Saichev and Sorenette created a range of p from 1 to 10^-3

    ρconstant = 1


    # defining the function 

    y = ((aconstant*nconstant*θconstant*ρconstant**(θconstant)*x**(-1-θconstant)) + (1-nconstant+aconstant*nconstant*ρconstant**(θconstant)*(x**-θconstant))**2) * np.exp(-(1-nconstant)*x-(aconstant*nconstant*(ρconstant**θconstant)/(1-θconstant))*x**(1-θconstant))
    
    # defining the widget function 

    # defining the power laws 

    y2 = x**(-1) / 30
    y3 = x**(-0.3) / 3
    y4 = ((a*n*θ*ρ**(θ)*x**(-1-θ)) + (1-n+a*n*ρ**(θ)*(x**-θ))**2) * np.exp(-(1-n)*x-(a*n*(ρ**θ)/(1-θ))*x**(1-θ))
    
    # changing size of the figure  

    plt.figure(figsize=(10, 10))

    # Labelling the functions plot

    plt.title("Saichev and Sorenette Equation 11")
    plt.xlabel("Values of  x")
    plt.ylabel("Values of f(x)")  

    # Changing the range 

    plt.ylim([0.00001, 100])
    plt.xlim([0.00001, 10])

    # Plotting the function

    plt.yscale("log")
    plt.xscale("log")
    plt.plot(x,y, color = 'b' ,label = "main function" )
    plt.plot(x,y2, '--' , color = 'g' ,label = "x^-1")
    plt.plot(x,y3, '--' , color = 'm' ,label = "x^-0.3")
    plt.plot(x,y4, color = 'orange' ,label = "widget function" )
    plt.legend() 
    plt.savefig("Saichev and Sornette Widget")
    plt.show()

widgets.interact(fn,A=(0,5,0.01),N=(0,5,0.01), Theta=(0,1,0.01), Rho=(0.001,1,0.001))

```
